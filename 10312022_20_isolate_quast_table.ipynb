```python
import pandas as pd
from pathlib import Path
all_quast_results_HLF = []
quast_results_directory = "/home/haley/Desktop/Complete_Assemblies_Plasmid/Hybrid_Ligation_Filtered/"
for quast_output in Path(quast_results_directory).glob('*_HLF_quast_output/transposed_report.tsv'):

    isolate_name = quast_output.parts[6]

    print(f"Parsing {quast_output}")
    isolate_quast_results = pd.read_csv(quast_output, sep='\t')
    isolate_quast_results['Isolate_Pathname'] = isolate_name

    all_quast_results_HLF.append(isolate_quast_results)

all_quast_results_HLF = pd.concat(all_quast_results_HLF)

print(all_quast_results_HLF)
```
